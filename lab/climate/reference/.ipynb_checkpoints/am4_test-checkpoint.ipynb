# Liby's Test on `am4`

```python
%matplotlib inline 
from mpl_toolkits.basemap import Basemap # import Basemap matplotlib toolkit
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset, num2date  # netcdf4-python module
```

```python
ds = Dataset('../data/GFDL_am4_cli_2009_01.nc')
ds
```

```python
plevs = ds.variables["plev"]
plevs
```

```python
plevs[:]
```

```python
clis = ds.variables["cli"]
clis
```

```python
cli_at_plev_150 = clis[-8, :, :]
print(cli_at_plev_150)
print(cli_at_plev_150.shape)
maximum = cli_at_plev_150.max()
minimum = cli_at_plev_150.min()
print("max: %f, min: %f" % (maximum, minimum))
```

```python
lats = ds.variables['lat'][:]
lons = ds.variables['lon'][:]
lons, lats = np.meshgrid(lons - 180, lats)
```

```python
fig = plt.figure(figsize=(12, 12))
m = Basemap()
m.drawcoastlines()
x, y = m(lons, lats)
cs = m.contourf(x, y, cli_at_plev_150, np.arange(minimum, maximum, 0.000001), cmap=plt.cm.jet)
cb = m.colorbar(cs)
```
